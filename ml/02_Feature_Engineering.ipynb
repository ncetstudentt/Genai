# Feature Engineering for Machine Learning
## Creating Powerful Features for Better Model Performance

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import (StandardScaler, MinMaxScaler, PolynomialFeatures,
                                   LabelEncoder, OneHotEncoder)
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

## 1. Feature Scaling

```python
# Create sample data
X = np.array([[1, 100], [2, 200], [3, 300], [4, 400]])

# Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# MinMax Scaling
minmax = MinMaxScaler()
X_minmax = minmax.fit_transform(X)

print('Original:\n', X)
print('\nStandardized:\n', X_scaled)
print('\nMinMax Scaled:\n', X_minmax)
```

## 2. Polynomial Features

```python
X_sample = np.array([[1, 2], [2, 3], [3, 4]])

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X_sample)

print('Original shape:', X_sample.shape)
print('Polynomial features shape:', X_poly.shape)
print('\nOriginal:\n', X_sample)
print('\nPolynomial Features:\n', X_poly)
```

## 3. Categorical Encoding

```python
# Label Encoding
colors = np.array(['red', 'green', 'blue', 'red', 'green']).reshape(-1, 1)

le = LabelEncoder()
colors_encoded = le.fit_transform(colors.ravel())

print('Original:', colors.ravel())
print('Label Encoded:', colors_encoded)

# One-Hot Encoding
ohe = OneHotEncoder(sparse_output=False)
colors_onehot = ohe.fit_transform(colors)

print('\nOne-Hot Encoded:')
print(colors_onehot)
```

## 4. Feature Selection

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=20, n_informative=5, random_state=42)

# SelectKBest
selector = SelectKBest(f_classif, k=5)
X_selected = selector.fit_transform(X, y)

print(f'Original shape: {X.shape}')
print(f'Selected features shape: {X_selected.shape}')
print(f'Selected feature indices: {selector.get_support(indices=True)}')
```

## 5. Dimensionality Reduction with PCA

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print(f'Original shape: {X.shape}')
print(f'PCA reduced shape: {X_pca.shape}')
print(f'Explained variance ratio: {pca.explained_variance_ratio_}')
print(f'Total variance explained: {pca.explained_variance_ratio_.sum():.2%}')
```
