# Advanced Data Analysis with Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample dataset
np.random.seed(42)
dates = pd.date_range('2023-01-01', periods=100)
sales_data = {
    'Date': dates,
    'Product': np.random.choice(['A', 'B', 'C'], 100),
    'Sales': np.random.randint(1000, 5000, 100),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 100)
}
df = pd.DataFrame(sales_data)
print(df.head())
```

## 1. Time Series Analysis

```python
# Set date as index
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

# Rolling average
df['Sales_MA'] = df['Sales'].rolling(window=7).mean()
print(df.head(10))
```

## 2. Pivot Tables

```python
# Create pivot table
pivot = df.pivot_table(values='Sales', index='Product', columns='Region', aggfunc='sum')
print(pivot)
```

## 3. Merging DataFrames

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['A', 'B', 'C']})
df2 = pd.DataFrame({'ID': [1, 2, 3], 'Score': [90, 85, 95]})

merged = pd.merge(df1, df2, on='ID')
print(merged)
```

## 4. Visualization

```python
df_reset = df.reset_index()
df_grouped = df_reset.groupby('Product')['Sales'].sum()

df_grouped.plot(kind='bar', figsize=(10, 6))
plt.title('Total Sales by Product')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```
